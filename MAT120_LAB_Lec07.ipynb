[Euler Method](https://en.wikipedia.org/wiki/Euler_method#:~:text=The%20Euler%20method%20is%20a,proportional%20to%20the%20step%20size.)



$dy/dx=y$

$y(0)=1$

```python
import numpy as np
import matplotlib.pyplot as plt

a=0 # lower boundary
b=3 # upper boundary
N=1001   # Step size
h=(b-a)/(N-1)

x=np.linspace(a,b,N)
y=0*x
y[0]=1 #initial value
for i in range(1,N):
  # prediction using Euler method
  y[i]=y[i-1]+h*y[i-1]

plt.plot(x,y)
plt.grid()
```

#$dy/dx=e^xsin(x+y)$
#$y(0)=0$

```python
f= lambda p,q: np.exp(p)*np.sin(p+q)

a=0
b=5
N=10001
h=(b-a)/(N-1)

x=np.linspace(a,b,N)
y=0*x
y[0]=0
for i in range(1,N):
  y[i]=y[i-1]+h*f(x[i-1],y[i-1])

plt.plot(x,y)
plt.grid()
```

#$d^2x/dt^2+16x=sin(3t)$
#$x(0)=2$
#$x'(0)=0$

```python
f= lambda p,q: np.sin(3*p)-16*q

a=0
b=5
N=10001
h=(b-a)/(N-1)
t=np.linspace(a,b,N)
v=t*0
x=t*0
x[0]=2
v[0]=0
for i in range(1,N):
  # At first predict the first order term
  v[i]=v[i-1]+h*f(t[i-1],x[i-1])
  # Now predict the main function
  x[i]=x[i-1]+h*v[i] # Check this should be v[i] not v[i-1]
plt.plot(t,x)
plt.grid()
```

# Projectile motion
#$x'=3$
#$y'=4-10t$
#$x(0)=0$
#$y(0)=0$

```python
a=0
b=0.8
N=1001
h=(b-a)/(N-1)
t=np.linspace(a,b,N)
x=0*t
y=0*t
x[0]=0
y[0]=0
for i in range(1,N):
  x[i]=x[i-1]+h*3
  y[i]=y[i-1]+h*(4-10*t[i-1])

plt.plot(x,y)
plt.grid()
```
