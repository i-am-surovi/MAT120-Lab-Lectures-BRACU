#Numerical Integration
Trapezoidal Rule

#$\int_{0}^{1}e^xdx$

```python
# Analytical Solution

import sympy as sp
x=sp.symbols('x')
sp.integrate(sp.exp(x),(x,0,1)).evalf()
```

```python
import numpy as np
def f(x):
  return np.exp(x)

# Trapezoid Rule

N= 1001
a=0
b = 1
h= (b-a)/(N-1)

s = 0
for i in range(1,N+1):
  s += f(i*h)*h

print(s)
```

#$\int_{1}^{2}e^xdx$

```python
# Analytical Solution

import sympy as sp
x=sp.symbols('x')
sp.integrate(sp.exp(x),(x,1,2)).evalf()
```

```python
def f(x):
  return np.exp(x)

# Trapezoid Rule

N= 1001
a=1
b = 2
h= (b-a)/(N-1)

s = 0
for i in range(1,N+1):
  # Predict the function by f(a+i*h) when not starts from 0
  s += f(a+i*h)*h

print(s)
```

#$\int_{0}^{1}e^xdx$

```python
from numpy.random import *
from numpy import *
import matplotlib.pyplot as plt

N = 9999 #large value will give more accurate result costing more computation time

x = rand(N)
temp = rand(N)
y = [e*i for i in temp]

plt.scatter(x,y)

z = [e**i for i in x]

plt.scatter(x,z)
n = 0
for i in range(N):
  if y[i] <= z[i]:
    n = n + 1

print(e*n/N)
```

#$\int_{0}^{1}sin(x)dx$

```python
# Analytical solution

import sympy as sp
x=sp.symbols('x')
sp.integrate(sp.sin(x),(x,0,1)).evalf()
```

```python
# Monte-Carlo simulation
a = 0
b = 1
N = 1000
xrand = rand(N)

def func(x):
    return np.sin(x)

sum = 0
for i in range(len(xrand)):
    sum += func(xrand[i])

integral = (b - a) * sum / N
print(integral)
```
