# Answer to the question no 1

```python
# Importing necessary library
import sympy as sp

# Writing the expression
x=sp.symbols('x')
y=x**4-100*x**2+10*x-2

# Plot the function within the interval
sp.plot(y,(x,-8,8))
```

The are total three extreme points of the function the within given interval. Two of them are local minima and other one is local maxima.

$Number \ of \ maxima=1$

$Number \ of \ minima=2$

# Answer to the question no 2

We already define the expression $y$ to answer the first question.

```python
# Let's see the expression again
display(y)
```

Let's plot the function within the range for more visualization

```python
sp.plot(y,(x,-1,1))
```

From the plot it is confirmed that there must be a mixima of the function within the given interval. Let's see how sympy can help us to find that maxima.

```python
# first derivative
diff1=sp.diff(y,x,1)

# Second derivative
diff2=sp.diff(y,x,2)

# solution of first derivative
soln=sp.solve(diff1)
display(soln)
```

The solve function didn't help us that much. However, we can ignore imaginary parts of each of the three solutions it provides. Because we already know that the function has at least 1 maxima and 2 minima(total 3 extrema).

```python
# Check for negligible imaginary parts and extract the real solutions
real_soln=[]
for i,s in enumerate(soln):
    if sp.im(s) < 1e-15:
        if sp.re(s)<0:
          s = -sp.sqrt(sp.im(s)**2+sp.re(s)**2)
        else:
          s = sp.sqrt(sp.im(s)**2+sp.re(s)**2)
    real_soln.append(s.evalf())
real_soln
```

```python

# Finding the maximum and minimum value
if len(real_soln)>0:
 for i in range(len(real_soln)):
   # Checking solutions within range 0 to 1
   if real_soln[i]>=-1 and real_soln[i]<=1:
        # Checking if real_soln[i] corresponds to maxima
        if diff2.subs(x,real_soln[i])<0:
              max_pt=real_soln[i]
              # Evaluating the maximum value
              max_val=y.subs(x,real_soln[i])
              print(f'The maximum value of the function at {max_pt} is {max_val}')
        # Checking if real_soln[i] corresponds to minima
        elif diff2.subs(x,real_soln[i])>0:
              min_pt=real_soln[i]
              # Evaluating the minimum value
              min_val=y.subs(x,real_soln[i]).evalf()
              print(f'The mimimum value of the function at {min_pt} is {min_val}')
else:
  print("No maxima or minima of the function within the given range.")
```

# Answer to the question no 3

Let's first find the maxima and minima of $y=sin(x)$ within the interval of $2$ to $4$

```python
# Declaring symbols
x=sp.symbols('x')

# Writing the expression
y=sp.sin(x)

# first derivative
diff1=sp.diff(y,x,1)
# Second derivative
diff2=sp.diff(y,x,2)

# solution of first derivative within interval
soln=sp.solveset(diff1,x,sp.Interval(2*sp.pi,4*sp.pi))
# Changing the solution set into a list
soln1=list(soln)

# Finding the maximum and minimum value
if len(soln1)>0:
  for k in range(len(soln1)):
    # Checking imaginary solutions
    if sp.im(soln1[k])!=0:
       continue
    # Checking if soln1[k] corresponds to maxima
    elif diff2.subs(x,soln1[k])<0:
      max_pt=soln1[k]
      # Evaluating the maximum value
      max_val=y.subs(x,soln1[k]).evalf()
      print(f'The maximum value of the function at {max_pt} is {max_val}')
    # Checking if soln1[k] corresponds to manima
    elif diff2.subs(x,soln1[k])>0:
      min_pt=soln1[k]
      # Evaluating the minimum value
      min_val=y.subs(x,soln1[k]).evalf()
      print(f'The mimimum value of the function at {min_pt} is {min_val}')
else:
  print("No maxima or minima")
```

Now we want to plot the function in the given interval and indicate the maxima and minima with red star marker.

```python
# Plotting markers corresponding to maxima and minima
marker=[]
if len(soln1)>0:
  for k in range(len(soln1)):
    X=soln1[k]
    Y=y.subs(x,soln1[k])
    marker.append({'args': [X,Y,'r*']})
p1=sp.plot(y,(x,2*sp.pi,4*sp.pi),markers=marker,show=False)
p1.show()
```
