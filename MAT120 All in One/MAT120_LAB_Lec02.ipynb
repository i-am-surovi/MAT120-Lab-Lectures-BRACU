#Maxima Minima

```python
# Writing expression f
f=x**3+10*x**2-40*x+5
```

```python
f
```

```python
# Plotting expression f
p3=sp.plot(f,title=' Graph of function',legend='x**3+10*x**2-40*x+5')
```

```python
# Getting first derivative of expression f
diff1=sp.diff(f,x,1)
diff1
```

```python
# Plotting first derivative of f
p4=sp.plot(diff1,title='first derivative',line_color='red')
```

```python
p3.extend(p4)
p3.show()
```

```python
# Solving first derivative
soln=sp.solve(diff1)
soln
```

```python
# Finding second derivative of expression f
diff2=sp.diff(f,x,2)
diff2
```

```python
max_pt=[]
min_pt=[]
# Iterating over the solution of first derivative
for i in range(len(soln)):
  if sp.im(soln[i])!=0:
    # Checking whether there is a maxima or minima for each solution
    if diff2.subs(x,soln[i]).evalf()<0:
       max_pt.append(soln[i])
    elif diff2.subs(x,soln[0]).evalf()>0:
       min_pt.append(soln[i])
```

```python
# List of x values where the main function have maxima
max_pt
```

```python
# List of x values where the main function have minima
min_pt
```

```python
# Substituting and evaluating the maximum value
maxima=f.subs(x,max_pt)
maxima.evalf()
```

```python
# Substituting and evaluating the minimum value of the function
minima=f.subs(x,min_pt)
minima.evalf()
```
